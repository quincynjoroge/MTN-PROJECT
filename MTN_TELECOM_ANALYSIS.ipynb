# Renaming columns

```python
#Renaming column PRODUTC TO PRODUCT AND DATETIME TO DATE_TIME in dataset 1
df3.rename(columns = {'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}, inplace = True)
df3
```

```python
#Renaming column DW_A_NUMBER TO DW_A_NUMBER_INT in dataset 2
df4.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace = True)
df4
```

```python
#Renaming column SIET_ID TO SITE_ID AND CELLID TO CELL_ID in dataset 3
df5.rename(columns = {'SIET_ID':'SITE_ID','CELLID':'CELL_ID'}, inplace = True)
df5
```

```python
#Renaming column SIET_ID TO SITE_ID AND CELLID TO CELL_ID in dataset 3
df6.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace = True)
df6
```

# Cleaning duplicates

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 1
df3.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 2
df4.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 3
df5.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates cell geo dataset
df6.drop_duplicates(inplace = True)
```

```python
#Merge CDR data sets
df = pd.concat([df3, df4,df5], ignore_index=True)
df
```

```python
#JOIN data sets CDR to Cell gee on the Cell_id and site id
df7 = df.merge(df6, on=["CELL_ID","SITE_ID"],left_index = True,right_index = True)
df7.head()
```

```python
#dropping columns 
df8 = df7.drop(['DW_A_NUMBER_INT','DW_B_NUMBER_INT','COUNTRY_A','COUNTRY_B','Unnamed: 0','LONGITUDE','LATITUDE'], axis=1)
df8
```

```python
# GETTING VALUES THAT ARE ONLY IN SERVICE
a = df8["STATUS"]=="In Service"
df8.where(a,inplace = True)

df8
```

```python
# Filtering the column value and cities and status

df_1= df8[['PRODUCT','VALUE','VILLES','STATUS']]

df_1
```
