# Dropping rows with null values

```python
#Removing rows with null values
#Table dataset 
df3.dropna(inplace = True)
df3
```

```python
#Removing rows with null values
#Table dataset 2
df4.dropna(inplace = True)
df4
```

```python
#Removing rows with null values
#Table dataset 3
df5.dropna(inplace = True)
df5
```

```python
#Removing rows with null values
#Cell geo dataset
df6.dropna(inplace = True)
df6

```

# Renaming columns

```python
#Renaming column PRODUTC TO PRODUCT AND DATETIME TO DATE_TIME in dataset 1
df3.rename(columns = {'PRODUTC': 'PRODUCT', 'DATETIME': 'DATE_TIME'}, inplace = True)
df3
```

```python
#Renaming column DW_A_NUMBER TO DW_A_NUMBER_INT in dataset 2
df4.rename(columns = {'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER':'DW_B_NUMBER_INT'}, inplace = True)
df4
```

```python
#Renaming column SIET_ID TO SITE_ID AND CELLID TO CELL_ID in dataset 3
df5.rename(columns = {'SIET_ID':'SITE_ID','CELLID':'CELL_ID'}, inplace = True)
df5
```

```python
#Renaming column SIET_ID TO SITE_ID AND CELLID TO CELL_ID in dataset 3
df6.rename(columns = {'SITE_CODE':'SITE_ID'}, inplace = True)
df6
```

# Cleaning duplicates

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 1
df3.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 2
df4.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates dataset 3
df5.drop_duplicates(inplace = True)
```

```python
#LOOKING FOR DUPLICATES
#Removing duplicates cell geo dataset
df6.drop_duplicates(inplace = True)
```

```python
#Merge CDR data sets
df = pd.concat([df3, df4,df5], ignore_index=True)
df
```

```python
#JOIN data sets CDR to Cell gee on the Cell_id and site id
df7 = df.merge(df6, left_index = True,right_index = True)
df7.head()
```

```python
#merge cell id columns
df7["CELL_ID"] = df7["CELL_ID_x"] + df7["CELL_ID_y"]
df7
```

```python
# GETTING VALUES THAT ARE ONLY IN SERVICE
a = df7["STATUS"]=="In Service"
df7.where(a,inplace = True)

df7 
```

```python
# Filtering the column value and cities and status

df_1= df7[['PRODUCT','VALUE','VILLES','STATUS']]

df_1
```

# Analysis

```python
#Which ones were the most used city for the three days?
#df_1.sort_values("VALUE", ascending=False).iloc[0:3]     
x = df_1.groupby(['VILLES']).count()
x.sort_values("VALUE", ascending=False).iloc[0:3] 
```

```python
#Which cities were the most used during business and home hours
df7_hours = df7[(df7.DATE_TIME >= "09.0.0") & (df7.DATE_TIME <= "17.0.0") ]
df7_hours[['VILLES','DATE_TIME']]
```

```python
# Most used city for the three days
x = df_1.groupby(['VILLES']).count()
x.sort_values("VALUE", ascending=False).iloc[0:1] 
```
